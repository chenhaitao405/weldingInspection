```python
DAGSHUB_REPO_OWNER = "chenhaitao405"
DAGSHUB_USER = "chenhaitao405"
DAGSHUB_REPO = "HFtest"
DAGSHUB_REPO_BRANCH = "main"
DAGSHUB_FULL_REPO = DAGSHUB_REPO_OWNER + "/" + DAGSHUB_REPO
DATASOURCE_NAME = "coco8seg"
DATASOURCE_PATH = "s3://HFtest/coco8seg"
```

```python
from dagshub.data_engine import datasources
```

```python
def get_or_create_datasource(name):
    try:
        ds = datasources.get_datasource(repo=DAGSHUB_FULL_REPO, name=name)
    except:
        ds = datasources.create(repo=DAGSHUB_FULL_REPO, name=name, path=DATASOURCE_PATH)
    return ds


ds = get_or_create_datasource(DATASOURCE_NAME)
```

```python
labeled = ds[ds['imported_annotations'].is_not_null()]
len(labeled.all())
```

```python
from data_utils.dataset_prepare import DataFunctions
dataset_func = DataFunctions(local_yolo_dir='generated/yolo_data', label_type='segmentation')
```

## Add Split Field to Dataset
Now we'll add a 'split' field to the dataset to distinguish between train, valid, and test sets.

```python
# Add split field to the labeled dataset
# You can adjust the ratios as needed (they should sum to 1.0)
split_stats = dataset_func.add_split_field(
    datasource=labeled,
    train_ratio=0.7,
    valid_ratio=0.15,
    test_ratio=0.15,
    random_seed=42  # Use a fixed seed for reproducibility
)

# Display the split statistics
print("\nSplit Statistics:")
for key, value in split_stats.items():
    print(f"{key}: {value}")
```

```python
# Verify that the split field has been added correctly
verification_stats = dataset_func.verify_split_field(labeled)

print("\nVerification completed!")
```
