将 YOLO 默认的目录结构映射为 repository 数据集中的实际路径

YOLO 标准目录结构假设图像存储在 `images/train` 或 `images/val` 目录下，

但本数据集图像直接存储在 `train/val` 目录中。
**转换示例：**
 - `images/train/000000000009.jpg` → `train/000000000009.jpg`
- `images/val/000000000025.jpg` → `val/000000000025.jpg`

```python
def remap_path(annotation_path):

    return annotation_path.replace("images\\", "").replace("images/", "").replace("\\", "/")
```

```python
ds.import_annotations_from_files(
  annotation_type="yolo", # or 'coco'
  path="E:\\CODE\\datasets\\coco8\\coco8-seg.yaml",  #本地路径
  field="imported_annotations", # name of the imported field
  yolo_type="segmentation", # or bbox
  remapping_function=remap_path
)
```
